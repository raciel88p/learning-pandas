```python
import numpy as np
dic = { #las varibles no pueden tener numeros y creando una tabla con variedad de datos y tiene que tener la misma cantidad de valores
    "edad": [10, 12, 14, 16, 18, 58],
    "Cm":   [155, 163, 183, 189, 98, 85],
    "pais": ["co", "pa", "cr", "ni", "US", "Ir"],
    "genero":["M", "F", "M", "F", "M", "M"],
    "Q1":[5, 10, 8, np.nan, 7, 8,],
    "Q2":[7, 9, 9, 8, 8, 8]

} 
dic
```

```python
df = pd.DataFrame(dic, index = ["Juan", "Edgar" ,"Flaco", "Carolina", "Tef", "Magaly"]) #creando una tabla con datos
df
```

```python
df.index #obteniendo el index de la tabla
```

```python
df.columns #obteniendo las columnas de la tabla
```

```python
df.values #obteniendo los valores de la tabla
```

```python
#obteniendo solo unos cuants datos
df["edad"]
```

```python
#llamando varios datos a la misma vez
df[["edad", "genero", "Q1"]]
```

```python
#localizando solo unos estudiantes
df.loc["Juan", ["edad", "genero", "Q1"]]
```

```python
#locaciando 2 estudiantes en la lista de datos
df.loc[["Juan", "Flaco"], ["edad", "genero", "Q1"]]
```

```python
#solo obtener un valor de la tabla
df.loc["Carolina", "Q1"]
```

```python
#obtener datos en base a condiciones
df["edad"] >= 12
```

```python
#obteniendo todos los datos mayores a 12 a;os
df[df["edad"] >= 12]
```

```python
#implementando mas filtros
df[(df["edad"] >= 12)&(df["pais"] == "cr")]
```

```python
#otra manera de usar filtros
df.query("edad >= 12")
```

```python
#conociendo cual tuvo mayor calificacion  y con los valores reales y no  true and false
df[df["Q2"] >= df["Q1"]]
```
