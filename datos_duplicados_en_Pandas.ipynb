```python
import pandas as pd
```

```python
df = pd.DataFrame({'a': ['w'] * 4 + ['x'] * 3 + ['y'] * 2 + ['z']+['v'], 
                   'b': [1, 1, 1, 1, 2, 2, 2, 3, 3, 4,5]})
df
```

```python
#le pedimos a la tabla que nos brinde todos los datos duplicados
df.duplicated()
```

```python
# keep='first' para marcar solo la primera ocurrencia
df.duplicated(keep='first')
```

```python
df.duplicated(keep='last')
```

```python
#podemos filtrar la data
df[~ df.duplicated()] #tiene el elmento de la negacion ~
```

```python
#simpre podes usar filtros para ver los duplicados
df.duplicated(keep=False)
```

```python
#usamos drop_duplicates' para eliminar los duplicados
df.drop_duplicates()
```

```python
#borrando los duplicados solo borrando los datos de una columna
df.drop_duplicates(['a'],keep='last')
```
