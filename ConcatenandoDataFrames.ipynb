```python
#Generando numeros aleatorios con numpy
numero_random1 = np.random.rand(2 , 5)*10
#randn genere numeros negativos y positivos
numero_random1
```

```python
numero_random2  = np.random.rand(2 , 5)* -1
numero_random2
```

```python
#numpy puede concatenar matrices
np.concatenate([numero_random1, numero_random2]) #concatenando matrices con una lista
```

```python
np.concatenate([numero_random1, numero_random2]).shape
```

```python
#concatenado tambi√©n al eje de las columnas
np.concatenate([numero_random1, numero_random2], axis=1)
```

```python
#viendo las dimenciones del objeto anterior
np.concatenate([numero_random1, numero_random2], axis=1).shape
```

```python
#haciendo todo lo anterior con pandas de concatenacion de vectores y matrices con pandas
variable_pandas = pd.Series(numero_random1[0], numero_random2[1])
variable_pandas
```

```python
#crear una nueva variable para hacer lo mismo con pandas
serie1 = pd.Series(numero_random1[0], index=['a', 'b', 'c', 'd', 'e'])
#generaos una seria con 5 elementos
serie1 
```

```python
#creando otra seria de elementos 
serie2 = pd.Series(numero_random2[0], index=['c', 'g', 'h', 'e', 'j'])
#generaos una seria con 5 elementos
serie2
```

```python
#concatenando los series 1  y 2
pd.concat([serie1, serie2], axis=1) #con axis cambiamos el comportamiento
# se van a concatenar a lo largo de las filas
```

```python
#como en las serie1 y serie1 se repiten los indice por eso da el error de NaN
#si el objetivo es concatenar 2 series a lo largo de los ejes de las columnas sin respetar el indice lo que tenemos que hacer es resetear el indice
serie1.reset_index(drop=True) #drop = true hace que los indices salgan enumerados del 0 al 4
#reiniciamos el indice previamente definido
```

```python
#creamos de nuevo nuestra concatenacion
pd.concat([serie1.reset_index(drop=True), serie2.reset_index(drop=True)], axis=1)
#nueva concatenacion de ambas variables sin respetar el indice
```

```python
#concatenando data frames 
#creamos el data frame a continuacion
data_1 = pd.DataFrame(np.random.rand(3, 2)*10, columns=['a','b'])
data_1
```

```python
data_2 = pd.DataFrame(np.random.rand(3, 2)*-1, columns=['a','b'], index=['2', '3', '4'])
data_2
```

```python
#concatenando un data frame en pandas
pd.concat([data_1, data_2])
```

```python
#Concatenando a lo largo de las columnas
pd.concat([data_1, data_2], axis=1)
```

```python
#haciendo la concatenacion  de 2 data frames pero haciendo solo enfasis a 1 columna
pd.concat([data_1, data_2], axis=1, join='inner')
#le indice que compar los 2 data frames
```

```python
#concatenar 2 data frames sin respetar los indices
pd.concat([
           data_1.reset_index(drop=True), 
           data_2.reset_index(drop=True)
           ], axis=1)
```

```python
#uniendo 2 data frames con la funcion append
data_1.append(data_2).append(data_2)
```

```python
#haciendo un append a lo largo de las comlumnas
#utilizamos una traspuesta
data_1.T
#invertir donde esa las columnas y row
```

```python
data_1.T.append(data_2.T).T
```
