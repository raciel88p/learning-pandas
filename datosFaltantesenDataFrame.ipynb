```python
#creando un dataFrame
df = pd.DataFrame(np.arange(0, 15).reshape(5, 3), columns=['a', 'b', 'c'])
df
```

```python
#anadiendo variables no definidas
df['d'] = np.nan
df['e'] = np.arange(15, 20)
df.loc[5,:] = pd.NA
df.loc[4,'a'] = pd.NA
df.loc[0,'d'] = 1
df.loc[5,'d'] = 10
df
```

```python
#para reconocer cuando un objeto es nulo usamos:
df.isnull()
```

```python
df.isna() #otra funcion similar a la anterior
```

```python
#Conocer el número de variables nulas por columna puede hacerse juntando el
#comando anterior con la funcion de suma:
df.isnull().sum()
```

```python
#Si lo que nos interesa es conocer el número de filas con elementos nulos, basta
#con usar axis=1:
df.isnull().sum(axis=1)
```

```python
#Conocer todos los elementos nulos de nuestro DataFrame:
df.size-df.isnull().sum().sum()
```

```python
#conocer estos elementos nos puede ayudar a filtrar el dataframe y comprender mas nuestros datos
df[df['a'].notnull()]
```

```python
#Con la funcion dropna es facil eliminar elementos faltantes
df.dropna()
```

```python
#cambiando el elemento NaN por un cero con la funcion fillnan()
df.fillna(0)
```

```python
#podemos remplazar valores de NaN con method="ffill":
df.fillna(method="ffill")
```

```python
#Si quisieramos remplazar con el valor previo usamos method="bfill":
df.fillna(method="bfill")
```

```python
#referente a lo anterior con otro metodo
df.fillna(method="bfill",axis=1)
```

```python
# también una serie para reemplazar los valores de una columna en especifico pero es importante que esten emparejadas
fill = pd.Series([100, 101, 102])
fill
```

```python
df['d'] = df['d'].fillna(fill)
df['d'] 
```

```python
#formas más usadas para reemplazar datos es usar el promedio de las columnas, esto se hace con la función mean. O si se quiere un mejor estimador, usamos median.
df.fillna(df.median())
```

```python
#Pandas también puede interporlar los valores faltanes calculando el valor que puede haber existido en el medio.
df_d = pd.concat([df[['d']], df[['d']].interpolate()],axis=1)
df_d.columns = ['d_antes','d_interpolado']
df_d
```
