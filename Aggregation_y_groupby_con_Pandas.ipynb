```python
#creando un nueva columna que se llama porcentaje de propina
df['prc_tips'] = df['tip']/df['total_bill']
df
```

```python
#vamos a volver a sacar el promedio del sexo
df.groupby('sex').mean()
#como en la columna anterior se creo un prc_tips 
#los mean son estimadores sesgados estadisticamente
```

```python
#se puede aplicar un grupo de funciones
df.groupby('sex')[['total_bill', 'prc_tips']].describe()
#obtnemos los parametros descriptivos de la columna tips y por genero con 2 indices
```

```python
#aplicando una funcion directamente
def mean_euro2dolars(x):
  return np.mean(x) * 1.12
```

```python
mean_euro2dolars(100)
```

```python
#podemos aplicar en el datafrmae anterior la funcion que escribimos
df.groupby('sex')[['total_bill', 'prc_tips']].apply(mean_euro2dolars)
#Hemos aplicado una nueva funcion que multiplica los datos de la tabla con 1.12 y transforma nuestras facturas a dolares
```

```python
#podemos agrupar usando mas variables
df.groupby(['sex', 'day', 'smoker'])[['total_bill', 'prc_tips']].apply(mean_euro2dolars)
```

```python
#podemos aplicar funciones definidas por nosotros o aplicar funciones definidas por numpy
df.groupby(['sex', 'day', 'smoker'])[['total_bill', 'prc_tips']].apply(np.std)
#calculamos la desbiacion estandar de la tabla anterior
```

```python
#podemos definir nuetras propias funciones dentro de la llamada del mismo dataframe
df.groupby(['sex', 'day', 'smoker'])[['total_bill', 'prc_tips']].apply(lambda x: np.mean(x)*1.12)
#me gusta mas llamar la funcion por aparte
```

```python
#aplicando mas de una funcion usamo aggregate
df.groupby(['sex', 'day', 'smoker'])[['total_bill', 'prc_tips']].aggregate([np.mean, np.max])
```

```python
#aplicando funciones en una sola categoria y funcion igual con agg
df.groupby(['sex', 'day', 'smoker'])[['total_bill', 'prc_tips']].agg([np.mean, np.max])
```

```python
#definiendo un dictionario, con un conjunto de funcion de tips min y maximo con factura total de un prodio y de dolares de un platillo
dict_agg = {'tip': [min, max], 'total_bill': [np.mean, mean_euro2dolars]}
dict_agg
```

```python
df.groupby(['sex', 'day', 'smoker'])[['total_bill', 'tip']].agg(dict_agg)
```

```python
#creando una columna que contiene numeros 1
df['ones'] = 1
df
```

```python
#vamos hacer grupos por sexo y fumadores
df_g = df.groupby(['sex', 'smoker'])[['ones']].sum()
df_g
#hemos echo un  grupo de quienes son sumadores entre sexos y quie no y sumado la todalidad de dicha grupo
```

```python
#convertir la tabla anterior en porcentajes
df_g.groupby(level=0).apply(
    
    lambda x:
    x / x.sum() * 100
)

#tranformar nuestra variables numericas a %
```

```python
#podemos transformar nuestras variables numericas en una variable categorica
df['bin_total'] = pd.cut(df['total_bill'], bins=[3, 18, 35, 60])
#cada uno de los varoles previamente definido en la tabla, va a quedar en una categoria
df
#como podemos ver cada uno de los valores se le asocia una categoria dentro de los valores 3, 18, 36 y 60
```

```python
#definiendo el ancho de las categorias utilizamos
pd.cut(df['total_bill'], bins=[3, 18, 35, 60]).value_counts()
```

```python
#sobre esa categoria podemos hacer groupby
df.groupby(['time', 'bin_total'])[['ones']].count()
#encontramos como  se  distribuye el tiempo del almuerzo y las comidas
```

```python
#conviertiendoloe en %
df.groupby(['time', 'bin_total'])[['ones']].count().apply(
    
    lambda x:
    x / x.sum() * 100
)
#nos dic que el 18% de las facturas entran en las facturas entre 3 y 18
```

```python
#aprendiendo a estruturar mejor los data frames con pivot y melt
#aprendiendo a crear tablas dinamicas con pivot table
#nos permite extrar datos categoricos e informacion de nuestro data frame
```

```python
#traemos el data frame
df.groupby(['sex','time'])[['total_bill']].mean()
```

```python
df_gp2 = df.groupby(['sex','time'])[['total_bill']].mean().reset_index()
df_gp2
```

```python
#creamos un tabla dinamica o un pivot table
df_gp2.pivot_table(values='total_bill', index='sex', columns='time' )
# a cambiado la table el sexo es el indice y las columnnas son los valores de tiempo
```

```python
#esto permite generar operaciones estadisticas aca
```

```python
df.pivot_table(values='total_bill', index='sex', columns='time' )
```

```python
#quiere decir que adentro hay una funcion que podes cambaiar o manipular segun las necesidades
df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=np.median )
#la funcion nos permite: calcular la media
```

```python
#Tambien podemos agregar un diccionario o lista de funciones dentro del aggfunc la media y la deviacion estandar
df_pivot = df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=[np.median, np.std])
df_pivot
```

```python
#vamos a deshacer las categorias del dataFrame
df_pivot.unstack()
#las columnas pasan hacer filas
```

```python
#haciendo lo anterior un dataframe con
df_pivot.unstack().reset_index()
```
