```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
#importamos una libreria de visualizacion
import seaborn as sns
df = sns.load_dataset('diamonds')
df
#base de datos de diamantes
```

```python
#aplicando la funcion group by
df.groupby('cut')
#crea un elemento de grupo
```

```python
#Aca podemos aplicar funciones para obtener el promedio
df.groupby('cut').mean()
```

```python
#otros estimadore estadisticos que no esta sesgado por un liar
df.groupby('cut').median()
```

```python
#aplicar los datos anteriores a nivel de columnas
df.groupby('cut')['carat'].median()
```

```python
#elegir varias funciones sobre el contar los datos que tenemos en la columnas para cada una de las categorias
df.groupby('cut')['carat'].count()
```

```python
#extrayendo el valor minimo 
df.groupby('cut')['carat'].min()
```

```python
#el groupby son como paquetes que corren diversas funciones
for i in [0,1,2,3,4]:
  print(i)
```

```python
"Asi es como funciona un groupby  como un for loop de python"
#agrupamos por un solo parametro
for key_group, group in df.groupby('cut'): #cada uno de los grupo que se ha hido formando dentro del grupo cut
  grouped_price = group['price'].mean() #se ido conformado por el precio y el prodio de la tabla

  print('Cut: {}, Price: {} \n'.format(key_group, grouped_price))
  #se aplican funcines a nivel de bucles
```

```python
#agrupamos por varios parametros
df.groupby(['cut', 'color'])['price'].mean().to_frame()
#to_frame generamos un data frame de 2 elementos conn multiples indices
```

```python
#podes crear nuestras propias funciones o pedier que se agregen nuevas funcionnes
# Utilizamos para eso agregate
df.groupby(['cut', 'color'])['price'].aggregate(['min',np.mean, max])
```

```python
#vamos a definir una funcion
def mean_kilo(x):
  return np.mean(x)/1000
```

```python
df.groupby(['cut', 'color'])['price'].aggregate(['min',np.mean, max, mean_kilo]).head(10)
#aplicamos la funcion mean_kilo al grupo by y solo mostramos 10 columunas con los kilos respectivamente
```

```python
#podemos hacer un diccionario de instrucciones que podemos agregar dentro de aggregate
#crear varios calculos estadisticos para diferentes categorias
dict_agg = {'carat': [min, max], 'price' :[np.mean, mean_kilo]}
dict_agg
#tiene como llave las columnas donde vamos aplicar nuestras funciones
```

```python
df.groupby(['cut', 'color']).aggregate(dict_agg).head(10)
#agrupamos por el corte y color y con la variable dicionario  y las llaves llamamos el min y el maximo y por otro lado el promedio y el kilo
```

```python
#otra funcionalidad que podemos crear es crear filtros
def f_filtro(x):
  return mean_kilo(x['price']) > 4

```

```python
df.groupby('cut').filter(f_filtro)
```
