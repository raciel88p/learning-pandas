```python
#definiendo 2 data frames 
#aca tenemos un dicionario dentro hay una lista de elementos
   
dataframe_izq = pd.DataFrame (
     {'x':['x0','x1','x2','x3'],
      'y':['y0','y1','y2','y3'],
      'w':['w0','w1','w2','w3']},
      #'Mix':['y2','y3','a2','a3']}, quitamos esta seccion para probar los otros formas
    index = [0, 1, 2, 3])

dataframe_izq
```

```python
dataframe_der = pd.DataFrame ( #aca tenemos un dicionario dentro hay una lista de elementos
    {'a':['a2','b3','c4','d5'],
    'b':['b2','b3','b4','b5'],
    'y':['y2','y3','y4','y5']},
    index = [2, 3, 4, 5])

dataframe_der
```

```python
#funcion de merge de pandas
pd.merge(dataframe_izq, dataframe_der)
```

```python
#se tiene que especificar el tipo de union que se hace
pd.merge(dataframe_izq, dataframe_der, how='inner', on = 'y')
```

```python
#uniendo la columna y con la columna mix
pd.merge(dataframe_izq, dataframe_der, how='inner', left_on = 'mix', right_on='y')
# se unifica por la columna  mix a la izquierda con la columna y por la derecha
```

```python
#unir los data frame por el caso contrario
pd.merge(dataframe_izq, dataframe_der, how='inner', left_on = 'Mix', right_on='a')
```

```python
#cantindad de merge que podemos tener
"""
how = 'inner' trae los parametros que tienen en comun los data frames
how = 'outer' une todos los datos de los 2 data frames
how = 'left' une los datos pero se respeta el data frame de la izquierda
how = 'rigth' une los datos pero se respeta el data frame de la derecha
"""
```

```python
#primer merge con inner
pd.merge(dataframe_izq, dataframe_der, how='inner', left_on = 'y', right_on='y')
```

```python
pd.merge(dataframe_izq, dataframe_der, how='left', left_on = 'y', right_on='y')
#se mantine el data frame de la iznquierda pero se anaden el data set de la derecha
```

```python
pd.merge(dataframe_izq, dataframe_der, how='right', left_on = 'y', right_on='y')
#se mantiene el data frame de la derecha y solo se agregan los datos de la izquirda
```

```python
#mantener tanto los datos de la izquierda como los de la derecha
pd.merge(dataframe_izq, dataframe_der, how='outer', left_on = 'y', right_on='y')
```
